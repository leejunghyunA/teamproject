```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
import seaborn as sns
from matplotlib import font_manager, rc
import warnings 
warnings.filterwarnings('ignore')
from scipy.special import erf
```

```python
df = pd.read_excel('월별 자동차신규등록건수.xlsx', index_col=0)
df
```

- 비율은 승합차비율
- 전기차 or 친환경차 비율 뽑기(국토교통부 자료)

```python
# 승합차만 뽑기
total_df = df.copy()
total_df['2015년'] = total_df['2015년']*0.83
total_df['2016년'] = total_df['2016년']*0.83
total_df['2017년'] = total_df['2017년']*0.83
total_df['2018년'] = total_df['2018년']*0.83
total_df['2019년'] = total_df['2019년']*0.83
total_df['2020년'] = total_df['2020년']*0.84
total_df['2021년'] = total_df['2021년']*0.83
total_df['2022년'] = total_df['2022년']*0.83

total_df
```

```python
# 합계뽑기
total_df1 = total_df.iloc[:-1, :-1]

total_df2 = pd.DataFrame(columns = total_df1.columns.tolist(), index=pd.Index(['전체sum']))
for i in total_df2.columns.tolist():
    total_df2.loc['전체sum', i] = total_df1[i].sum()
total_df2
```

```python
# 승합차의 비율 뽑기
# 친환경차 비율(14~15년 후는 친환경차 보유고객이 전기차 고객일 확률 높음)
df1 = df.copy()
df1['2015년'] = df['2015년']*0.83*0.09
df1['2016년'] = df['2016년']*0.83*0.11
df1['2017년'] = df['2017년']*0.83*0.15
df1['2018년'] = df['2018년']*0.83*0.2
df1['2019년'] = df['2019년']*0.83*0.25
df1['2020년'] = df['2020년']*0.84*0.34
df1['2021년'] = df['2021년']*0.83*0.47
df1['2022년'] = df['2022년']*0.83*0.64

df1
```

```python
df1 = df1.iloc[:-1, :-1]
```

```python
df1.columns.tolist()
```

```python
# 친환경차, 승용차 비중
# 합계만 뽑기
total1 = pd.DataFrame(columns = df1.columns.tolist(), index=pd.Index(['친환경차sum']))
for i in df1.columns.tolist():
    total1.loc['친환경차sum', i] = df1[i].sum()
print('친환경차, 승합차 합계')
total1
```

```python
# 전기차 비율만
df2 = df.copy()
df2['2015년'] = df['2015년']*0.83*0.001
df2['2016년'] = df['2016년']*0.83*0.01
df2['2017년'] = df['2017년']*0.83*0.01
df2['2018년'] = df['2018년']*0.83*0.02
df2['2019년'] = df['2019년']*0.83*0.04
df2['2020년'] = df['2020년']*0.84*0.06
df2['2021년'] = df['2021년']*0.83*0.09
df2['2022년'] = df['2022년']*0.83*0.11
df2
```

```python
df2 = df2.iloc[:-1, :-1]
```

```python
# 전기차, 승용차 비중
# 합계만 뽑기
total2 = pd.DataFrame(columns = df2.columns.tolist(), index=pd.Index(['전기차sum']))
for i in df2.columns.tolist():
    total2.loc['전기차sum', i] = df2[i].sum()
print('=================승용차, 전기차 합계==================')
total2
```

```python
total = pd.concat([total1, total2, total_df2])
total
```

```python
total.to_excel('서울시년도별전체자동차수.xlsx')
```

- 2020년도의 전기차 -> 2010년도의 친환경차 건수 -> 2024년
- 2019년 -> 2009년 -> 2023년
