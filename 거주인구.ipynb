# 2021-12

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
stay_df = pd.read_excel('staypop202112.xlsx', header=None)
#정렬
stay_df=stay_df.rename(columns = {0:'행정기관코드',1:'행정기관', 2:'총인구수', 3:'세대수', 4:'세대당 인구', 5:'남자 인구수', 6:'여자 인구수', 7:'남여 비율'})
stay_df.drop([0,1,2,3],inplace=True)
stay_df.reset_index(drop=True,inplace=True)
stay_df.head()
```

## 필요컬럼 추가

- 시군구명

```python
import re
for i in range(len(stay_df)):
    name = re.search('\s{1}(\w+구)',stay_df.loc[i,'행정기관']).group(1)
    stay_df.loc[i,'시군구명'] = name
#     print(name)
stay_df.head()
```

- 일자

```python
stay_df['year'] = 2021
stay_df['month'] = 12
stay_df.head()
stay_df.columns
```

- 컬럼순서

```python
stay_df = stay_df.reindex(columns=['year', 'month','행정기관코드', '행정기관','시군구명', '총인구수', '세대수', '세대당 인구', '남자 인구수', '여자 인구수', '남여 비율'])
stay_df.head()
```

## 파일저장


```python
stay_df.to_excel('restaypop202112.xlsx',index=False)
```
