```python
import pandas as pd
```

```python
data = pd.read_excel("자치구별 연료별 자동차 등록현황.xlsx")
data.head()
```

## 컬럼 및 값 정리

```python
# 컬럼명 변경
data = data.drop(index=0, axis=0)
data.columns = ['등록연도', '사용본거지', '시군구코드', '용도', '차종종별', '연료', '건수']
data.head()
```

```python
# 등록연도 값 정리 (년도만 출력)
for i in range(3077):
    data.iloc[i, 0] = int(data.iloc[i, 0][:4])
    
data['등록연도']
```

```python
## 값 확인하기
print("****null값****", '\n', data.isnull().sum(), '\n')
print("****사용본거지****", '\n', data['사용본거지'].value_counts(), '\n')
print("****시군구코드****", '\n', data['시군구코드'].value_counts(), '\n')
print("****용도****", '\n', data['용도'].value_counts(), '\n')
print("****차종종별", '\n', data['차종종별'].value_counts(),'\n')
print("****연료", '\n', data["연료"].value_counts())
```

```python
# 문자 값 변경 방법: 인코딩 -LabelEncoder
#각 값 객체의 개수에 따라 입력해서 바꿔줌(0, 1 로 2개면 2개, 1,2,3으로 3개면 3개)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() #객체를 만들어줌
features = ['차종종별', '용도'] #바꿀 열 따로  뽑아줌(for 문 사용하기 위해)
for f in features:
    data[f] = le.fit_transform(data[f])  #fit _ transform
data.head()
```

```python
print(data['차종종별'].value_counts(), '\n')
data['용도'].value_counts()
```

```python
# 빈데이터프레임 생성
df = pd.DataFrame(columns=['년도', '주소', '용도', '연료', '건수'])
df
```

```python
# 연료별로 묶기 

# 2020년도
# 20년도 전기가 아니것 
df1 = data[(data['등록연도'] == 2020) & (data['연료']!="전기")]

# 20년도 전기가 아닌것
df2 = data[(data['등록연도'] == 2020) & (data['연료']=="전기")]
```

```python
# 전기가 아닌것
df1
```

```python
# 전기차일 때 
df2
```

```python
# 20년도 전기가 아니것 지역별로 묶기

df_1 = pd.pivot_table(df1,
                     values = '건수',
                     index = '사용본거지',
                     columns = '연료',
                     aggfunc = 'sum')
df_1
```
