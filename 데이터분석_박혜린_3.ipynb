### 2022년 1월 df : news_df_1.csv

```python
# 네이버 뉴스 살해 징역 검색 >> 2022/1월

# 라이브러리
import requests
from bs4 import BeautifulSoup 
import urllib
import urllib.request as rq
import pandas as pd
import requests
import openpyxl
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
import re
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager


# 웹드라이버 설정
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

# 빈 데이터 프레임 생성
news_df = pd.DataFrame()


while True:
    try:

        url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%82%B4%ED%95%B4%20%EC%A7%95%EC%97%AD&sort=0&photo=0&field=0&pd=3&ds=2022.01.01&de=2022.01.31&cluster_rank=26&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from20220101to20220131,a:all&'
        headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}

        # 페이지 2, 3, 4... 페이지로 넘기기 위해 params 이용
        for i in range(0, 100):
            response = requests.get(url, headers=headers, params = {'start' : 1+10*i})
            soup = BeautifulSoup(response.text, 'lxml')
        
            # 태그로 텍스트 찾은 후 데이터 프레임에 넣어줌
            news_title = [title['title'] for title in soup.find_all('a', attrs={'class':'news_tit'})]
            news_url = [url['href'] for url in soup.find_all('a',attrs={'class':'news_tit'})]
            dates = [date.get_text() for date in soup.find_all('span', attrs={'class':'info'})]
          
        
            news_date = []
            for date in dates:
                if re.search(r'\d+.\d+.\d+',date) !=None:
                    news_date.append(date)
        
            new_df = set(news_title)        

            df = pd.DataFrame({'date':news_date,'title':news_title, 'url':news_url})
            news_df  = pd.concat([news_df, df], ignore_index=True)
            #start += 10

    except:
        print('crawling done')
        break
```

```python
news_df_1 = news_df.copy()
news_df_1
```

```python
# 페이지 넘어갔는지 체크
news_df_1.head(30)
```

```python
# 중복 기사 제거
news_df_1_1 = news_df_1.drop_duplicates(['title'])
news_df_1_1.to_csv('news_df_1.csv')
news_df_1_1 = pd.read_csv('news_df_1.csv')
news_df_1_1
```
