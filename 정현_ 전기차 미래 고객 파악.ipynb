# <19년 전기차 신규등록 (소유 파악) - 종속변수>

```python
data2 = pd.read_csv("2019년 연료별 자동차 등록현황.csv", encoding = 'cp949')
# 19년 자치구별 연료별 자동차 등록현황(서울열린데이터광장)
data2
```

## 1. 컬럼 값 확인 및 정리

## 1-1) 컬럼 정리

```python
# 컬럼명 변경
for i in range(len(data2['년월'])):
    data2.iloc[i, 0] = int(data2.iloc[i, 0][:4])
    
data2['년월']
```

```python
# 컬럼명 변경
data2.columns = ['등록연도', '사용본거지', '연료', '용도', '승용', '승합', '화물', '특수', '합계']
data2.head()
```

```python
data2.fillna(0, inplace = True)
data2.head()
```

```python
# 값 확인
print(data2.isnull().sum(), '\n')
print(data2.dtypes, '\n')
print("****사용본거지****", '\n', data2['사용본거지'].value_counts(), '\n')
print("****연료", '\n', data2["연료"].value_counts(), '\n')
print("****용도****", '\n', data2['용도'].value_counts())
```

```python
len(data2)
```

```python
# 천단위 때문에 문자열 변경이 어려워 정규식으로 , 제거
data2['승용'] = data2['승용'].str.replace(',','')
data2['승합'] = data2['승합'].str.replace(',','')
data2['화물'] = data2['화물'].str.replace(',','')
data2['합계'] = data2['합계'].str.replace(',','')

# null 값 제거
data2.fillna(0, inplace = True)

# 문자로 변경
data2 = data2.astype({'승용':int,
                     '승합':int,
                     '화물':int,
                     '특수':int,
                     '합계':int})
data2.dtypes
```

## 1-2) 20, 21년도와 같은 형태로 만들기

```python
# 새로운 데이터프레임 형성
data3 = data2[['사용본거지', '연료', '용도']]
data3
```

```python
# 중복제거
dt = data3.drop_duplicates(['사용본거지', '연료', '용도'])
dt = dt.reset_index()
dt = dt.drop('index', axis= 1)
dt
```

```python
# 차종종별로 건수 추가하기

li1 = data2["사용본거지"].unique().tolist()

li2 = data2["연료"].unique().tolist()

# 용도 리스트
li3 = data2["용도"].unique().tolist()

# 승용
dt1 = dt.copy()
dt1['차종종별'] = '승용'
dt1['건수'] = 0

# 승합
dt2 = dt.copy()
dt2['차종종별'] = '승합'
dt2['건수'] = 0

# 화물
dt3 = dt.copy()
dt3['차종종별'] = '화물'
dt3['건수'] = 0

# 특수
dt4 = dt.copy()
dt4['차종종별'] = '특수'
dt4['건수'] = 0

# 해당 값 추가
for a in li1:
    for b in li2:
        for c in li3:
            dt1['건수'][(dt1['사용본거지']==a)&(dt1['연료']==b)&(dt1['용도']==c)] = data2['승용'][(data2['사용본거지']==a)&(data2['연료']==b)&(data2['용도']==c)].sum()
            dt2['건수'][(dt2['사용본거지']==a)&(dt2['연료']==b)&(dt2['용도']==c)] = data2['승용'][(data2['사용본거지']==a)&(data2['연료']==b)&(data2['용도']==c)].sum()
            dt3['건수'][(dt3['사용본거지']==a)&(dt3['연료']==b)&(dt3['용도']==c)] = data2['승용'][(data2['사용본거지']==a)&(data2['연료']==b)&(data2['용도']==c)].sum()
            dt4['건수'][(dt4['사용본거지']==a)&(dt4['연료']==b)&(dt4['용도']==c)] = data2['승용'][(data2['사용본거지']==a)&(data2['연료']==b)&(data2['용도']==c)].sum()
                    
# 데이터프레임 행방향 합치기
result = pd.concat([dt1,dt2,dt3,dt4], ignore_index=True)
result.head()
```

```python
result
```

```python
# 값 확인
print(result.isnull().sum(), '\n')
print(result.dtypes, '\n')
print("****사용본거지****", '\n', result['사용본거지'].value_counts(), '\n')
print("****연료", '\n', result["연료"].value_counts(), '\n')
print("****용도****", '\n', result['용도'].value_counts())
```

## 2. 지역별 묶기

### 1) 지역별로 건수 합

```python
df_19 = pd.pivot_table(result,
                     values = '건수',
                     index = '사용본거지',
                     columns = '연료',
                     aggfunc = 'sum')
df_19
```

```python
# null 값 확인

result[result['연료']=="하이브리드(CNG+전기)"]

# null 값 0으로 변경
df_19.fillna(0, inplace = True)

# int로 변경
df_19 = df_19.astype(int)
df_19.head()
```

### 2) 지역별로 차종종별 count

```python
df_19_type = pd.pivot_table(result,
                     values = '건수',
                     index = '사용본거지',
                     columns = ['차종종별','연료'],
                     aggfunc = 'sum')
df_19_type.head()
```

```python
# null 값 0으로 변경
df_19_type.fillna(0, inplace = True)

# int로 변경
df_19_type = df_19_type.astype(int)
df_19_type.head()
```

## 3. 파일 저장

```python
# 전체 19년 데이터
import json
df_19.to_json("total_car19.json", orient = 'columns')
# 차종별 20년 데이터
df_19_type.to_json("type_car19.json", orient = 'columns')
```
