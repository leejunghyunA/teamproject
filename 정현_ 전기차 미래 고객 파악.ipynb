```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("자치구별 연료별 자동차 등록현황.xlsx")
# 20_21년 자치구별 연료별 자동차 등록현황(서울열린데이터광장)
data.head()
```

## 1. 컬럼 값 확인 및 정리

## 1-1) 컬럼 정리

```python
# 컬럼명 변경
data = data.drop(index=0, axis=0)
data.columns = ['등록연도', '사용본거지', '시군구코드', '용도', '차종종별', '연료', '건수']
data.head()
```

```python
# 등록연도 값 정리 (년도만 출력)
for i in range(3077):
    data.iloc[i, 0] = int(data.iloc[i, 0][:4])
    
data['등록연도']
```

```python
## 값 확인하기
print("****null값****", '\n', data.isnull().sum(), '\n')
print("****사용본거지****", '\n', data['사용본거지'].value_counts(), '\n')
print("****시군구코드****", '\n', data['시군구코드'].value_counts(), '\n')
print("****용도****", '\n', data['용도'].value_counts(), '\n')
print("****차종종별", '\n', data['차종종별'].value_counts(),'\n')
print("****연료", '\n', data["연료"].value_counts())
```

## 2. 연도별 지역별로 건수 묶기

- 연도별로 묶기

```python
# 2020년도
df20 = data[(data['등록연도'] == 2020)]

# 21년도 전기가 아닌것
df21 = data[(data['등록연도'] == 2021)]
```

```python
# 2020년
df20
```

```python
# 2021년
df21
```

### 1) 지역별로 신규건수 묶기

```python
# 20년

df_20 = pd.pivot_table(df20,
                     values = '건수',
                     index = '사용본거지',
                     columns = '연료',
                     aggfunc = 'sum')
df_20
```

```python
# null 값 확인

df20[df20['연료']=="하이브리드(CNG+전기)"]

# null 값 0으로 변경
df_20.fillna(0, inplace = True)
df_20.head()
```

```python
# 21년
df_21 = pd.pivot_table(df21,
                     values = '건수',
                     index = '사용본거지',
                     columns = '연료',
                     aggfunc = 'sum')
df_21
```

```python
# null 값 0으로 변경
df_21.fillna(0, inplace = True)
df_21.head()
```

### 2) 지역별로 차종종별 count

```python
# 20년

df_20_type = pd.pivot_table(df20,
                     values = '건수',
                     index = '사용본거지',
                     columns = ['차종종별','연료'],
                     aggfunc = 'sum')
df_20_type.head()
```

```python
# null값 제거
df_20_type.fillna(0, inplace = True)
df_20_type.head()
```

```python
# 21년

df_21_type = pd.pivot_table(df21,
                     values = '건수',
                     index = '사용본거지',
                     columns = ['차종종별', '연료'],
                     aggfunc = 'sum')
df_21_type.head()
```

```python
# null값 제거
df_21_type.fillna(0, inplace = True)
df_21_type.head()
```

## 3. 파일 저장

```python
# json 파일로 저장

# 전체 20년 데이터
df_20.to_json("total_car20.json", orient = 'table')

# 전체 21년 데이터
df_21.to_json("total_car21.json", orient = 'table')

# 차종별 20년 데이터
df_20_type.to_json("type_car20.json", orient = 'table')

# 차종별 21년 데이터
df_21_type.to_json("type_car20.json", orient = 'table')
```
