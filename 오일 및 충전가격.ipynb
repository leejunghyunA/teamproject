# oil data

## 2019-2021 자료불러오기

```python
%pwd
```

```python
#불러오기
import pandas as pd
Y2019 = pd.read_csv('./data/유가/과거_판매가격(주유소)201901-201912.csv',encoding='cp949')
Y2019.drop([0],inplace=True)
Y2019.reset_index(drop=True, inplace=True)
print(len(Y2019))
Y2019.head()
```

```python
#불러오기
import pandas as pd
Y2020 = pd.read_csv('./data/유가/과거_판매가격(주유소)202001-202012.csv',encoding='cp949')
Y2020.drop([0],inplace=True)
Y2020.reset_index(drop=True, inplace=True)
print(len(Y2020))
Y2020.head()
```

```python
#불러오기
import pandas as pd
Y2021 = pd.read_csv('./data/유가/과거_판매가격(주유소)202101-202112.csv',encoding='cp949')
Y2021.drop([0],inplace=True)
Y2021.reset_index(drop=True, inplace=True)
print(len(Y2021))
Y2021.head()
```

- 합치기

```python
Yall = pd.concat([Y2019,Y2020,Y2021],axis=0)
Yall.reset_index(drop=True, inplace=True)
print(len(Yall)) # 482+474+442 = 1398
Yall #2021년 마지막 주유소가 8월까지밖에 정보가 없음.(오류아님!)
```

## 기간구분

```python
import re
for i in range(len(Yall)):
    y = Yall.loc[i,'기간'].split(" ")[0]
    Yall.loc[i,'year'] = re.search('(\d+)',y).group()
    m = Yall.loc[i,'기간'].split(" ")[1]
    Yall.loc[i,'month'] = re.search('(\d+)',m).group()
Yall
```

## 월평균 가격

- 연도별 그룹화

```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
# import matplotlib.font_manager as fm

#한글 폰트설정
# font_location = 'C:/cakd7/NanumGothic.ttf'
# font_name = fm.FontProperties(fname=font_location).get_name()
# plt.rc('font', family=font_name) # For Windows

oil_pivot = pd.pivot_table ( Yall,
                    index = ['year','month'],
                    values = '휘발유',
                    aggfunc = 'mean')
oil_pivot.plot(kind='bar')
```

```python
oil_pivot
```

## 파일저장
- 태블로 시각화 할거라서 그냥 필요컬럼만 뽑기

```python
oil_df = Yall[['year','month','지역','고급휘발유','휘발유','경유']] #실내등유 0 값 많아서 뺌
oil_df.head()
```

```python
oil_df.to_excel('oil_df.xlsx',index=None)
```
